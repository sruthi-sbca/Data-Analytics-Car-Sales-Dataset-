```python
#for reading files:
import pandas as pd
df = pd.read_csv('carsales.csv')
#display first five rows and columns
df.head()
```

```python
#display last 5 rows and columns
df.tail()
```

```python
#find missing values
df.isnull().sum()
```

```python
#drop those which are not targeted
df.dropna(inplace=True)
```

```python
#cheack if data is cleaned
df.isnull().sum()
```

```python
# duble-check for missing,duplicates,and display info and summary
print("Missing values:\n", df.isnull().sum())
print("\nNumber of duplicate rows:", df.duplicated().sum())
print("\nDataset info:")
print(df.info())
print("\nSummary statistics:")
print(df.describe())
```

```python
#accesing row by index
print(df.iloc[0])
```

```python
#check shape
df.shape
```

```python

import matplotlib.pyplot as plt
import seaborn as sns

# Histogram
plt.hist(df['Sales_in_thousands'], bins=20, color='blue', edgecolor='black')
plt.title('Histogram')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.show()
```

```python
#box plot
sns.boxplot(data=df, x='Sales_in_thousands', color='purple')
plt.title('Box Plot')
plt.show()
```

```python
#KDE plot
sns.kdeplot(df['Sales_in_thousands'], fill=True, color='orange')
plt.title('KDE Plot')
plt.xlabel('Values')
plt.ylabel('Density')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# ECDF for car prices
sns.ecdfplot(data=df, x='Price_in_thousands', color='blue')
plt.title('ECDF of Car Prices')
plt.xlabel('Price')
plt.ylabel('Cumulative Probability')
plt.grid(True)
plt.show()
```

```python
#rugplot
sns.rugplot(df['Sales_in_thousands'], height=0.1, color='red')
plt.title('Rug Plot')
plt.show()

```

```python
correlation_matrix = df.corr(numeric_only=True)  # Include only numeric columns

# Heatmap
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Heatmap of Correlations')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Scatter plot
sns.scatterplot(data=df, x='Sales_in_thousands', y='Price_in_thousands')
plt.title('Scatter Plot')
plt.show()
```

```python
#line plot
sns.lineplot(data=df, x='Sales_in_thousands', y='Price_in_thousands', hue='Manufacturer')
plt.title('Line Plot')
plt.show()
```

```python
#regression plot
sns.regplot(data=df, x='Sales_in_thousands', y='Price_in_thousands')
plt.title('Regression Plot')
plt.show()
```

```python
#violin plot
sns.violinplot(data=df, x='Sales_in_thousands', color='cyan')
plt.title('Violin Plot')
plt.show()
```

PROBLEM STATEMENT: FIND WHICH MODEL HAS THE HIGHEST SALES


```python
#finding max in all columns
df.max()
```

```python
#Find the max  in sale column
df['Sales_in_thousands'].max(axis=0)
```

```python
#find which model has the highest sale through boolean
df[df['Sales_in_thousands'] == df['Sales_in_thousands'].max()]
```

Conclusion:The car model with the highest sales was determined as " Ford F series, with a sales figure of 540.561 thousand units.
